```python
# prompt: Use Broadcast Variables to Display Movie Names Instead of ID
# Numbers

from pyspark.sql import SparkSession
from pyspark.sql.functions import broadcast

# Initialize SparkSession
spark = SparkSession.builder.appName("MovieRecommendations").getOrCreate()

# Sample movie data (replace with your actual data)
movie_data = [
    (1, "Movie A"),
    (2, "Movie B"),
    (3, "Movie C"),
]
movie_df = spark.createDataFrame(movie_data, ["movie_id", "movie_title"])

# Sample ratings data (replace with your actual data)
ratings_data = [
    (1, 1, 5),
    (2, 2, 4),
    (3, 3, 3),
]
ratings_df = spark.createDataFrame(ratings_data, ["user_id", "movie_id", "rating"])


# Create a broadcast variable from the movie DataFrame
broadcast_movies = broadcast(movie_df)

# Perform a join operation, using the broadcast variable
joined_df = ratings_df.join(broadcast_movies, "movie_id")

# Display the result (movie titles instead of movie IDs)
joined_df.select("user_id", "movie_title", "rating").show()

# Stop the SparkSession
spark.stop()
```
