```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import sum, col

spark = SparkSession.builder.appName("TotalSpentByCustomer").getOrCreate()

orders = spark.read.csv("path/to/orders.csv", header=True, inferSchema=True)

total_spent = orders.groupBy("customer_id") \
                   .agg(sum("order_amount").alias("total_spent"))

total_spent.show()

```
